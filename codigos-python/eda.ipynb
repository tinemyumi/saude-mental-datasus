# **Filtrando dados sobre saúde mental no Estado de São Paulo**

- Os códigos CID-10 para transtornos mentais e comportamentais variam de F00 a F99.
- Código para o estado de São Paulo = 35

```python
# Dados sobre saúde mental (CID-10 F00-F99)
df_saude_mental = df[df['DIAG_PRINC'].str.startswith('F', na=False)].copy()
# Internações do estado de São Paulo (código 35)
df_saude_mental = df_saude_mental[df_saude_mental['UF_ZI'].str.startswith('35', na=False)].copy()

# Colunas para analisar
colunas_analise = ['UF_ZI', 'ANO_CMPT', 'MES_CMPT', 'ESPEC', 'MUNIC_RES', 'DIAG_PRINC', 'CID_ASSO', 'CID_MORTE', 'DT_INTER', 'DT_SAIDA', 'DURACAO_INTERNACAO_DIAS', 'QT_DIARIAS', 'DIAS_PERM']

# Conversão para formato de data
df_saude_mental['DT_INTER'] = pd.to_datetime(df_saude_mental['DT_INTER'])
df_saude_mental['DT_SAIDA'] = pd.to_datetime(df_saude_mental['DT_SAIDA'])

# Conversão da duração da internação em dias
df_saude_mental['DURACAO_INTERNACAO_DIAS'] = (df_saude_mental['DT_SAIDA'] - df_saude_mental['DT_INTER']).dt.days

df_saude_mental[colunas_analise]

```

# **Filtrando dados de população apenas sobre o estado de São Paulo.**

O código para o estado de São Paulo é 35, então estamos filtrando todos os municípios que começam com 35.

- D1C = Município (código)
- A coluna D1C_6_DIGITOS foi adicionada para acrescentar os códigos dos municípios, excluindo o último dígito, uma vez que ele é o dígito verificador do IBGE e o DATASUS utiliza o formato com os 6 primeiros digítos.

```python
# Filtra o dataframe original do IBGE (df_ibge) para incluir as linhas onde o código da coluna 'D1C' começa com 35 que é o código de SP.
df_sp = df_ibge[df_ibge['D1C'].str.startswith('35', na=False)].copy()

# Nova coluna que recebe os primeiros 6 caracteres da coluna D1C
df_sp['D1C_6_DIGITOS'] = df_sp['D1C'].str[:6]

df_sp.head()
```

# **Análise Exploratória**

**Legenda:**
- **internacoes_psiquiatricas:** Número de internações psiquiátricas por município de residência.
- **internacoes_por_100k **= taxa de internações por 100.000 habitantes
-** Baixo Acesso (0 a 50 internações por 100k habitantes):** Representa os municípios com uma taxa muito baixa de internações. Isso pode sugerir que há pouca oferta de serviços hospitalares de saúde mental ou que os pacientes dessas cidades precisam se deslocar para outras.
- **Médio Acesso (50 a 150 internações por 100k habitantes):** Representa os municípios com uma taxa intermediária de internações.
- **Alto Acesso (150 ou mais internações por 100k habitantes):** Representa os municípios com uma taxa muito alta de internações. Isso pode indicar que essas cidades são grandes centros de saúde mental, que atendem tanto a sua própria população quanto pacientes de outras cidades (os que você chamou de "pacientes exportados").


```python
# Número de internações por município de residência
internacoes_por_municipio = df_saude_mental.groupby('MUNIC_RES').size().reset_index(name='internacoes_psiquiatricas')
print(internacoes_por_municipio)
```

```python
# Junção dos conjuntos de dados (internação e população)
df_completo = pd.merge(internacoes_por_municipio, df_sp, left_on='MUNIC_RES', right_on='D1C_6_DIGITOS', how='left')

# Conversão do formato da coluna V (Valor da População) para numérico
df_completo['V'] = pd.to_numeric(df_completo['V'], errors='coerce')

# Cálculo da taxa internações por 100k: nº internações de cada município / população normalizada para 100.000 habitantes
df_completo['internacoes_por_100k'] = (df_completo['internacoes_psiquiatricas'] / df_completo['V']) * 100000

bins = [0, 50, 150, float('inf')]
labels = ['Baixo Acesso', 'Médio Acesso', 'Alto Acesso']
df_completo['nivel_acesso'] = pd.cut(df_completo['internacoes_por_100k'], bins=bins, labels=labels, right=False)


print(df_completo['nivel_acesso'].value_counts())
```
