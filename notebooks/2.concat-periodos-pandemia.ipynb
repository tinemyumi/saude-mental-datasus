# **1. Pré Pandemia**

```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

arquivos = [
    '/content/drive/MyDrive/DadoDataSUS/dados_sih/dados_datasus/2014.parquet',
    '/content/drive/MyDrive/DadoDataSUS/dados_sih/dados_datasus/2015.parquet',
    '/content/drive/MyDrive/DadoDataSUS/dados_sih/dados_datasus/2016.parquet',
    '/content/drive/MyDrive/DadoDataSUS/dados_sih/dados_datasus/2017.parquet',
    '/content/drive/MyDrive/DadoDataSUS/dados_sih/dados_datasus/2018.parquet',
    '/content/drive/MyDrive/DadoDataSUS/dados_sih/dados_datasus/2019.parquet'
    ] # Colocar os arquivos que queremos juntar

output_path = '/content/drive/MyDrive/Dataframes/df_pre_pandemia.parquet'

writer = None

for arquivo in arquivos:
  table = pq.read_table(arquivo) # lê um por vez
  if writer is None:
    writer = pq.ParquetWriter(output_path, table.schema)
  writer.write_table(table)

writer.close()
```
