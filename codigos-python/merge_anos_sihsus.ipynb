```python
import pandas as pd

arquivos = [
    '/content/drive/MyDrive/Dataframes/2022.xlsx',
    '/content/drive/MyDrive/Dataframes/2023.xlsx',
    '/content/drive/MyDrive/Dataframes/2024.xlsx'
]

anos = [2022,2023,2024]

lista_df = []

for arquivo, ano in zip(arquivos, anos):
  df = pd.read_excel(arquivo)
  df['ANO'] = ano
  lista_df.append(df)

df_total = pd.concat(lista_df)

df_total['ANO_CMPT'].value_counts()
```

```python
df_total.to_excel('/content/drive/MyDrive/Dataframes/df_total.xlsx', index=False)
```
