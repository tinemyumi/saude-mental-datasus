# **1. Análise Inicial do Dataset Pré-Pandemia**

Data: 13/09/2025 Autor: Larissa Tinem

## **1.1. Filtrando dados sobre saúde mental no Estado de São Paulo e transformação de dados**

- Os códigos CID-10 para transtornos mentais e comportamentais variam de F00 a F99.
- Código para o estado de São Paulo = 35
- Transformação de tipos de dados

**Alterações**

- Data: 20/09/2025 Autor: Larissa Tinem
- Data: 08/10/2025 Autor: Larissa Tinem - Acréscimo de outras colunas
- Data: 27/01/2026 Autor: Larissa Tinem - Filtragem de colunas para df do período pré-pandemia

```python
# Carregando o arquivo em um dataframe
df_pre_pandemia = pd.read_parquet(
    '/content/drive/MyDrive/Dataframes/df_pre_pandemia.parquet',
    columns=[
      'N_AIH',
      'UF_ZI',
      'IDENT',
      'ANO_CMPT',
      'MES_CMPT',
      'NASC',
      'IDADE',
      'SEXO',
      'RACA_COR',
      'ESPEC',
      'MUNIC_RES',
      'CEP',
      'DIAG_PRINC',
      'DIAG_SECUN',
      'CAR_INT',
      'COBRANCA',
      'MUNIC_MOV',
      'CID_ASSO',
      'CID_MORTE',
      'COMPLEX',
      'DT_INTER',
      'DT_SAIDA',
      'QT_DIARIAS',
      'DIAS_PERM',
      'MORTE']
)
```

```python
# Dados sobre saúde mental (CID-10 F00-F99)
df_saude_mental_pre = df_pre_pandemia[df_pre_pandemia['DIAG_PRINC'].str.startswith('F', na=False)].copy()
# Internações do estado de São Paulo (código 35)
df_saude_mental_pre = df_saude_mental_pre[df_saude_mental_pre['UF_ZI'].str.startswith('35', na=False)].copy()
```

## **1.2. Transformação dos tipos de dados**

- Conversão das colunas **NASC, DT_INTER e DT_SAIDA** para formato data (aaaammdd)
- Conversão das colunas **ANO_CMPT, MES_CMPT, IDADE, QT_DIARIAS, DIAS_PERM** para números inteiros

Data: 20/09/2025 Autor: Larissa Tinem - Conversão de tipos de dados



```python
import importlib

# Transformação dos tipos de dados

# Conversão para formato de data
df_saude_mental_pre['DT_INTER'] = pd.to_datetime(df_saude_mental_pre['DT_INTER'])
df_saude_mental_pre['DT_SAIDA'] = pd.to_datetime(df_saude_mental_pre['DT_SAIDA'])
df_saude_mental_pre['NASC'] = pd.to_datetime(df_saude_mental_pre['NASC'])

# Conversão para int
conversao_int = ['ANO_CMPT', 'MES_CMPT', 'IDADE', 'QT_DIARIAS', 'DIAS_PERM']
for coluna in conversao_int:
    df_saude_mental_pre[coluna] = df_saude_mental_pre[coluna].astype(int)
```

```python
df_saude_mental_pre.head()
```

## **1.3. Salvando DF de saúde mental no Google Drive**

```python
df_saude_mental_pre.to_parquet('/content/drive/MyDrive/Dataframes/df_sih_sus_tratado/saude_mental_pre_pandemia.parquet')
```
