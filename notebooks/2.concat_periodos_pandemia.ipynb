# **1.  Concatenação dos arquivos do SIH-SUS**

```python
import pandas as pd
import pyarrow as pa
import pyarrow.parquet as pq

arquivos = [
    '/content/drive/MyDrive/Dataset/dados_sih/2015.parquet',
    '/content/drive/MyDrive/Dataset/dados_sih/2016.parquet',
    '/content/drive/MyDrive/Dataset/dados_sih/2017.parquet',
    '/content/drive/MyDrive/Dataset/dados_sih/2018.parquet',
    '/content/drive/MyDrive/Dataset/dados_sih/2019.parquet',
    '/content/drive/MyDrive/Dataset/dados_sih/2020.parquet',
    '/content/drive/MyDrive/Dataset/dados_sih/2021.parquet',
    '/content/drive/MyDrive/Dataset/dados_sih/2022.parquet',
    '/content/drive/MyDrive/Dataset/dados_sih/2023.parquet',
    '/content/drive/MyDrive/Dataset/dados_sih/2024.parquet',
    '/content/drive/MyDrive/Dataset/dados_sih/2025.parquet'
    ] # Colocar os arquivos que queremos juntar

output_path = '/content/drive/MyDrive/Dados/df_sihsus.parquet'

writer = None

for arquivo in arquivos:
  table = pq.read_table(arquivo) # lê um por vez
  if writer is None:
    writer = pq.ParquetWriter(output_path, table.schema)
  writer.write_table(table)

writer.close()
```
