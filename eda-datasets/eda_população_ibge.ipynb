```python
df = pd.read_csv("/content/drive/MyDrive/DadosPopulação/brazil_municipalities_population.csv")
```

```python
cols = ['year', 'id_municipality', 'name_municipality', 'abbrev_state', 'population']
df_sp = df[((df['year'] == 2022) & (df['abbrev_state'] == 'SP'))][cols].copy()
df_sp.head()
```

```python
# "Cortar" o último dígito verificador do IBGE para bater com os códigos de municípios no dataset do Datasus
df_sp['COD_MUN_6'] = df_sp['id_municipality'].astype(str).str[:-1]
df_sp['COD_MUN_6'] = df_sp['COD_MUN_6'].astype(int)
```
